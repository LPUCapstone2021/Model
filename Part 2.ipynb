```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
this_df = pd.read_csv('new_cars.csv')
this_df.head()
```

```python
def top_five_cars(car):
    
    clicked_car = car
    # default input
    desired_car_user_rating=this_df[clicked_car]
    # desired_car_user_rating.head()
    
    similar_to_selected_car=this_df.corrwith(desired_car_user_rating)
    corr_selected_car = pd.DataFrame(similar_to_selected_car, columns=['Correlation'])
    # corr_selected_car.dropna(inplace=True)
    # corr_selected_car.head()
    
    corr_selected_car.sort_values(by='Correlation', ascending=False).head(10)
    new_df = corr_selected_car.nlargest(10,'Correlation')
    # new_df.head()
    
    final_list = list(new_df.index)[:6]
    final_list.remove('Ford Aspire')
    return final_list
```

```python
# car_clicked = input(),          here input will be car clcked by user
# deafult input
car_clicked = 'Ford Aspire'
cars = top_five_cars(car_clicked)
print("You may also like : \n", cars)
```
