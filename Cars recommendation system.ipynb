```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv('cars_features.csv', index_col = 0)
df.head()
```

```python
df.drop("Power",  axis = 1, inplace = True)
df.head()
```

```python
# len(df)
type(df)
```

```python
for i in range(1,len(df)):
    print(df.loc[i])
    print("========================")
```

```python
# for i in range(1,len(df)):
#     print(list(df.loc[i]))
#     print("==============================")

user = [1, 1, 1, 1, 0, 0, 0, 1, 0, 0]
user_vector = np.array(user)

sd_list = []
for i in range(1,len(df)+1):
    v = list(df.loc[i])
    vector = np.array(v)
    dist = np.linalg.norm(vector - user_vector) 
    sd_list.append(round(dist, 2))
#     print(dist)
print(sd_list)
    
```

```python
df["dist"] = sd_list
df.head()
```

```python
df.sort_values('dist', inplace =  True)
df.head()
```

```python
similar_cars_id = []
for id in df.index: 
    similar_cars_id.append(id) 
print("Top 5 cars according to user's desired features : \n")
print(similar_cars_id[:5])
```

```python
user_features_priorities = [2.5, 1, 1.5, 2, 0, 0, 0, 0.5, 0, 0, 0]
# top_similar_cars_id = similar_cars_id[:5]
dict1 = {}
for i in similar_cars_id[:5]:
    score = np.dot(user_features_priorities, list(df.loc[i]))
    dict1[i] = score

print(dict1)
```

```python
sorted_values = sorted(dict1.values(), reverse = True) 
print(sorted_values)
```

```python
print("Top Five recommendation on the basis of users desired features : ")
def get_key(val):
    for key, value in dict1.items():
         if val == value:
             return key
 
for i in sorted_values:
    car_id = get_key(i)
    del dict1[car_id]
    print("Recommendation", str(i), " :", car_id )
     
```
