```python
recommeded_cars_ids = []
def get_key(val):
    for key, value in dict1.items():
         if val == value:
             return key
 
for i in sorted_scores:
    car_id = get_key(i)
    recommeded_cars_ids.append(car_id)
    del dict1[car_id]
    
print(recommeded_cars_ids)
```
