```python
this_df = pd.read_csv('new_cars.csv')
this_df.head()
```

```python
df1 = pd.DataFrame(this_df.melt(id_vars = ["Users"]))
df1.columns = ['Users', 'Cars name', 'Ratings']
df1.head()
```

```python
df1.shape
```

```python
df_cars = df1.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
).fillna(0)
```

```python
df_cars.head()
```

```python
# mean = np.nanmean(df_cars, axis=1)
# df_normalized = (df_cars.T - mean).T
# df_normalized
```

```python
from sklearn.metrics.pairwise import pairwise_distances

# Cars Similarity Matrix
car_correlation = 1 - pairwise_distances(df_cars, metric='cosine')
car_correlation[np.isnan(car_correlation)] = 0
print(car_correlation)
```

```python
car_correlation[car_correlation < 0] = 0
car_correlation
```

```python
car_correlation.shape
```

```python
df_car_correlation = pd.DataFrame(car_correlation)
df_car_correlation.columns = df_cars.index
df_car_correlation.head()
```

```python
final_df = df_car_correlation.T
final_df.columns = df_cars.index
final_df.head()
```

```python
final_df.loc['Maruti XL6'].sort_values(ascending=False)[:11]
```

# Evaluation

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df1, test_size=0.30, random_state=31)
```

```python
print(train.shape)
print(test.shape)
```

```python
temp_train = train.copy()
temp_test = test.copy()
```

```python
temp_train['Ratings'] = temp_train['Ratings'].apply(lambda x: 0 if x >= 0 else 1)
temp_test['Ratings'] = temp_test['Ratings'].apply(lambda x: 1 if x >= 0 else 0)
```

```python
# Cars not rated by user are marked 1 for later prediction. 
temp_train = temp_train.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
).fillna(1)

# Cars not rated by user are marked 0 for later evaluation. 
temp_test = temp_test.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
).fillna(0)
```

### Prediction for the cars rated by the user

```python
test_DF = test.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
)
```

```python
test_car_correlation = 1 - pairwise_distances(test_DF.fillna(0), metric = 'cosine')
# test_car_correlation[np.isnan(test_car_correlation)] = 0
test_car_correlation[test_car_correlation < 0] = 0
```

```python
test_car_predicted_ratings = np.dot(test_car_correlation, test_DF.fillna(0))
test_car_predicted_ratings = np.multiply(test_car_predicted_ratings, temp_test)
test_car_predicted_ratings.head()
```

```python
X2 = test_car_predicted_ratings.copy() 
X2 = X2[X2 > 0]

scaler = MinMaxScaler(feature_range = (1, 5))
scaler.fit(X2)
y2 = (scaler.transform(X2))
y2 = pd.DataFrame(y2)
y2.head()
```

```python
#to confirm
temp_test.head()
```

```python
test_original = test.pivot(
    index = 'Cars name',
    columns = 'Users',
    values = 'Ratings'
)
```

```python
from numpy import *
total_non_nan = np.count_nonzero(~np.isnan(y2))

# rmse = (sum(sum((test_original - y2)**2))/total_non_nan)**0.5
# print(rmse)
```

```python
import math

lista = []
for i in range(0,125):
    templist = [b / m for b,m in zip(list(y2.iloc[i].values),list(test_original.iloc[i].values))]
    cleanedList = [x for x in templist if (math.isnan(x) == False)]
    lista.append(sum(cleanedList))
#   len(lista)
```

```python
accuracy = (sum(lista)/total_non_nan)*100
print("Accuracy of this model is : {} %".format(np.round(accuracy, 2)))
```
