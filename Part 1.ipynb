```python
dict1 = {}
def make_dict(user_features_priorities):
    df.drop('E_distance', axis = 1, inplace = True)
#     dict1 = {}
    for i in similar_cars_id:
        score = np.dot(user_features_priorities, list(df.loc[i]))
        dict1[i] = score

    return dict1
```

```python
score_dict = make_dict(user_features_priorities_d)
print(score_dict)
```

```python
temp_dict = dict(dict1)
```

```python
recommeded_cars_ids = []
def produce_ids():
    sorted_scores = sorted(score_dict.values(), reverse = True) 
    
    def get_key(val):
        for key, value in score_dict.items():
             if val == value:
                 return key

    for i in sorted_scores:
        car_id = get_key(i)
        recommeded_cars_ids.append(car_id)
        del score_dict[car_id]

    return recommeded_cars_ids
```

```python
recom_cars_id = produce_ids()
print(recom_cars_id)
```
